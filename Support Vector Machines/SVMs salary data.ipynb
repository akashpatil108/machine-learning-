```python
import pandas as pd
import numpy as np
```

```python
sd_train=pd.read_csv("SalaryData_Train.csv")
sd_train.sample(n=10)
```

```python
sd_train.info()
```

```python
sd_train.isnull().sum()
```

```python
col=sd_train.select_dtypes(include=object).columns
col
```

```python

sd_train['workclass']=sd_train['workclass'].astype('category')
sd_train['education']=sd_train['education'].astype('category')
sd_train['maritalstatus']=sd_train['maritalstatus'].astype('category')
sd_train['occupation']=sd_train['occupation'].astype('category')
sd_train['relationship']=sd_train['relationship'].astype('category')
sd_train['race']=sd_train['race'].astype('category')
sd_train['native']=sd_train['native'].astype('category')
sd_train['sex']=sd_train['sex'].astype('category')
```

```python
sd_train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder().fit_transform
```

```python
sd_train['Salary']=le(sd_train["Salary"])
```

```python
sd_train.Salary.head()
```

```python
col=sd_train.select_dtypes(include='category').columns
col
```

```python
for i in col:
    sd_train[i]=le(sd_train[i])
sd_train.sample(n=10)
```

```python
# Splitting the data into x and y as input and output
x=sd_train.iloc[:,:13]
y=sd_train.iloc[:,13]
```

```python
x.sample(n=5)
```

```python
y.sample(n=5)
```

```python
sd_train.Salary.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
X_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(X_train)
x_test=scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
clf=SVC()
clf.fit(x_train,y_train)
pred=clf.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print('accuracy_score', accuracy_score(pred,y_test))
print('classification_report\n', classification_report(pred,y_test))
confusion_matrix(pred,y_test)
```
